# Áreas de limpeza predial

# Lista de Areas cadastradas

```python
# retorna todas as áreas cadastradas, que o usário tem permissão para visualizar
# codigo de permissão necessário --> 252: Pode visualizar áreas de limpeza predial
response = requests.get("http://127.0.0.1:9000/api/areas/ListAreasLimpezaPredial/", headers=headers)
response.json()['results']
df = pd.DataFrame(response.json()['results'])
display(df)
```

```python
# Lista de servicos que podem ser enviados pelo usuario logado
# nao precisa de permissao
print("Servicos disponiveis")
response = requests.get("http://127.0.0.1:9000/api/catalogo_de_servicos/ListCatalogodeServicoLimpezaPredialFromForms/", headers=headers)
response.json()
df = pd.DataFrame(response.json()['results'])
display(df)


# Lista de localidades que podem ser enviados pelo usuario logado
# nao precisa de permissao
print("Localidades disponiveis")
response = requests.get("http://127.0.0.1:9000/api/localidades/ListLocalidadeLimpezaPredialFromForms/", headers=headers)
response.json()
df = pd.DataFrame(response.json()['results'])
display(df)
```

# Atualizar area

```python
# as chaves body da atualização é exatamente igual o body da criacao
# atualiando a periodicidade para bimestral
body = {
    'nome': "area jupyter", 
    'dimensao': 1568, 
    'servico': 1, # vem de /api/catalogo_de_servicos/ListCatalogodeServicoJardinagemFromForms/
    'localidade': 1,  # vem de api/localidades/ListLocalidadeJardinagemFromForms/
    'foto': "", #Tipo file    
}
```

```python
# essa rota recebe o id random da área como parametro de url, consultar em em api/areas/ListAreasLimpezaPredial/
# atualizando a área criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 251: Pode editar áreas de limpeza predial
response = requests.put("http://127.0.0.1:9000/api/areas/AreasLimpezaPredialUpdate/IeXYawWDHyda/", body, headers=headers)
response.json()
```

# Alterar status area

```python
# essa rota recebe o id random da área como parametro de url, consultar em api/areas/ListAreasLimpezaPredial/
# atualizando a área criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 
# 254: Pode desmobilizar áreas de limpeza predial ou demosbilizar (Desmobilizado) ou
# 255: Pode reabilitar áreas de limpeza predial para reabilitar (Mobilizado)

body = {
    'status': 'Desmobilizado',
}

response = requests.put("http://127.0.0.1:9000/api/areas/AreaLimpezaPredialAlterStatus/IeXYawWDHyda/", body, headers=headers)
response.json()
```

# Se deletar area

```python
# essa rota recebe o id random da área como parametro de url, consultar em api/areas/ListAreasLimpezaPredial/
# codigo de permissão necessário --> 253: Pode excluir áreas de limpeza predial

response = requests.get("http://127.0.0.1:9000/api/areas/IfDeleteAreasLimpezaPredial/IeXYawWDHyda/", headers=headers)
response.json()
```

# Deleção da área

```python
# essa rota recebe o id random da área como parametro de url, consultar em api/areas/ListAreasLimpezaPredial/
# codigo de permissão necessário --> 253: Pode excluir áreas de limpeza predial

response = requests.delete("http://127.0.0.1:9000/api/areas/DeleteAreasLimpezaPredial/IeXYawWDHyda/", headers=headers)
response.json()
```
