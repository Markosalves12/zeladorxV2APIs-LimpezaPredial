# atualizar Localidade

```python
# as chaves body da atualização é exatamente igual o body da criacao
# atualiando a nome para localidade do jupyter atualziado
body = {
    'nome': "localidade do jupyter atualziado", 
    'lat_med': 14.56,
    'long_med': 8.64, 
    'unidade': 5, # unidade disponivel em ListUnidadeFromForms
}
```

```python
# essa rota recebe o id random da localidade como parametro de url, consultar em api/areas/ListLocalidadeLimpezaPredial/
# atualizando a localidade criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 291: Pode editar localidades
response = requests.put("http://127.0.0.1:9000/api/localidades/LocalidadeLimpezaPredialUpdate/E8wjtPNovVKT/", body, headers=headers)
response.json()
```

# Alterar Status da localidade

```python
# essa rota recebe o id random da localidade como parametro de url, consultar em api/localidades/ListLocalidadeLimpezaPredial/
# atualizando a localidade criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 
# 294: Pode desmobilizar localidades para demosbilizar (Desmobilizado) ou
# 295: Pode reabilitar localidades para reabilitar (Mobilizado)

body = {
    'status': 'Desmobilizado'
}

response = requests.put("http://127.0.0.1:9000/api/localidades/LocalidadeLimpezaPredialAlterStatus/E8wjtPNovVKT/", body, headers=headers)
response.json()
```

# Se deletar localidade

```python
# essa rota recebe o id random da localidade como parametro de url, consultar em api/localidades/ListLocalidadeLimpezaPredial/
# codigo de permissão necessário --> 293: Pode excluir localidades
response = requests.get("http://127.0.0.1:9000/api/localidades/IfDeleteLocalidadeLimpezaPredial/E8wjtPNovVKT/", headers=headers)
response.json()
```

# Deleção da localidade

```python
# essa rota recebe o id random da localidade como parametro de url, consultar em api/localidades/ListLocalidadeLimpezaPredial/
# codigo de permissão necessário --> 293: Pode excluir localidades
response = requests.delete("http://127.0.0.1:9000/api/localidades/DeleteLocalidadeLimpezaPredial/E8wjtPNovVKT/", headers=headers)
response.json()
```
