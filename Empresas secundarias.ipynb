## Criação de novas empresas secundarias de limpeza predial

```python
body = {
    'nome': 'empresa jupyter',
    'setor': 2, # vem de api/zeladorx/MyMacroServices/
    'empresaprimaria': 1, #empresa primaria precisa estar validada em http://127.0.0.1:9000/api/zeladorx/EmpresaPrimaria/
}
```

```python
response = requests.post("http://127.0.0.1:9000/api/empresas_secundarias/CreateEmpresaSecundariaLimpezaPredial/", body, headers=headers)
response.json()
```

# Detalhes da empresa secundaria

```python
response = requests.get("http://127.0.0.1:9000/api/empresas_secundarias/EmpresaSecundariaLimpezaPredialDetail/ePwoYsT1ngJx/", body, headers=headers)
response.json()
```

# Atualização da empresa secundaria

```python
new_body = {
 'nome': 'EmpJardPostMaN edita jupyter',
 'setor': [1, 2], # vem de api/zeladorx/MyMacroServices/
 'empresaprimaria': 1,  #empresa primaria precisa estar validada em http://127.0.0.1:9000/api/zeladorx/EmpresaPrimaria/
}
```

```python
# esse end point recebe o id random da empreswa secundaria como parametro api/empresas_secundarias/EmpresaSecundariaLimpezaPredialUpdate/id_random/
response = requests.put("http://127.0.0.1:9000/api/empresas_secundarias/EmpresaSecundariaLimpezaPredialUpdate/ePwoYsT1ngJx/", new_body, headers=headers)
response.json()
```

# Alterar status da empresa secundaria

```python
new_body = {
    'status': 'Desmobilizado'
}
```

```python
response = requests.put("http://127.0.0.1:9000/api/empresas_secundarias/EmpresaSecundariaLimpezaPredialAlterStatus/ePwoYsT1ngJx/", new_body, headers=headers)
response.json()
```

# Se deletar o objeto

```python
response = requests.get("http://127.0.0.1:9000/api/empresas_secundarias/IfDeleteEmpresaSecundariaLimpezaPredial/ePwoYsT1ngJx/", new_body, headers=headers)
response.json()
```

# Deleção empresa secundaria

```python
response = requests.delete("http://127.0.0.1:9000/api/empresas_secundarias/DeleteEmpresaSecundariaLimpezaPredial/ePwoYsT1ngJx/", headers=headers)
response.json()
```
