# Catalogo de serviços

```python
# retorna todas os servicos cadastradas, que o usário tem permissão para visualizar
# codigo de permissão necessário --> 262: Pode visualizar serviços do catálogo
response = requests.get("http://127.0.0.1:9000/api/catalogo_de_servicos/ListCatalogodeServicoLimpezaPredial/", headers=headers)
response.json()['results']
df = pd.DataFrame(response.json()['results'])
display(df)
```

# Atualizar serviço catalogo

```python
# as chaves body da atualização é exatamente igual o body da criacao
# atualiando o nome para catalogo teste
body = {
    'nome': 'catalogo teste',
    'EmpresaSecundaria': 9, #vem de api/empresas_secundarias/ListEmpresaSecundariaLimpezaPredialFromForms/
}
```

```python
# essa rota recebe o id random da servico como parametro de url, consultar em api/catalogo_de_servicos/ListCatalogodeServicoLimpezaPredial/
# atualizando a servico criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 261: Pode editar serviços do catálogo
response = requests.put("http://127.0.0.1:9000/api/catalogo_de_servicos/CatalogoServicosLimpezaPredialUpdate/WPb30c0lAeaF/", body, headers=headers)
response.json()
```

# Alterar status catalogo servico

```python
# essa rota recebe o id random da servico como parametro de url, consultar em api/catalogo_de_servicos/ListCatalogodeServicoLimpezaPredial/
# atualizando a servico criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 
# 264: Pode desmobilizar serviços do catálogo para demosbilizar (Desmobilizado) ou
# 265: Pode reabilitar serviços do catálogo para reabilitar (Mobilizado)

body = {
    'status': 'Mobilizado',
}

response = requests.put("http://127.0.0.1:9000/api/catalogo_de_servicos/CatalogoServicosLimpezaPredialAlterStatus/WPb30c0lAeaF/", body, headers=headers)
response.json()
```

# Se deletar servico catalogo

```python
# essa rota recebe o id random da servico como parametro de url, consultar em api/catalogo_de_servicos/ListCatalogodeServicoLimpezaPredial/
# codigo de permissão necessário --> '263: Pode excluir serviços do catálogo'
response = requests.get("http://127.0.0.1:9000/api/catalogo_de_servicos/IfDeleteServicoCatalogoLimpezaPredial/WPb30c0lAeaF/", headers=headers)
response.json()
```

# Deleção do serviço

```python
# essa rota recebe o id random da servico como parametro de url, consultar em api/catalogo_de_servicos/ListCatalogodeServicoJardinagem/
# codigo de permissão necessário --> '263: Pode excluir serviços do catálogo'
response = requests.delete("http://127.0.0.1:9000/api/catalogo_de_servicos/DeleteServicoCatalogoLimpezaPredial/WPb30c0lAeaF/", headers=headers)
response.json()
```
